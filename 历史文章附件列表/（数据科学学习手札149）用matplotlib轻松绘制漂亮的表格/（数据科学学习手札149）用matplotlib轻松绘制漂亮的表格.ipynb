# 2 基于plottable快速绘制漂亮的表格

## 2.1 从简单例子出发

```python
import numpy as np
import pandas as pd
from plottable import Table
import matplotlib.pyplot as plt

# 构造示例数据表格
demo_df = pd.DataFrame(
    np.random.rand(5, 4),
    columns=[f'field{i}' for i in range(1, 5)]
)
demo_df = demo_df.round(3)

demo_df
```

```python
# 在matplotlib画布基础上渲染目标数据框的表格图
fig, ax = plt.subplots()
Table(demo_df)

plt.show()
```

## 2.2 plottable的常用方法

### 2.2.1 控制表格奇数偶数行底色

```python
fig, ax = plt.subplots()
Table(demo_df, odd_row_color='#f8f9fa', even_row_color='white')

plt.show()
```

### 2.2.2 控制表头单元格与数据单元格样式

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    col_label_cell_kw={
        'facecolor': '#a5d8ff'
    },
    cell_kw={
        'facecolor': '#e7f5ff'
    }
)

plt.show()
```

### 2.2.3 调节单元格文字样式

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    col_label_cell_kw={
        'facecolor': '#a5d8ff'
    },
    cell_kw={
        'facecolor': '#e7f5ff'
    },
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman',
        'ha': 'center' # 水平居中
    }
)

plt.show()
```

### 2.2.4 配置行分割线

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman',
        'ha': 'center' # 水平居中
    },
    col_label_divider=False, # 默认为True
    footer_divider=True, # 默认为False
    row_dividers=False # 默认为True
)

plt.show()
```

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman',
        'ha': 'center' # 水平居中
    },
    col_label_divider_kw={
        'linestyle': '--'
    },
    row_divider_kw={
        'color': '#20c997'
    }
)

plt.show()
```

### 2.2.5 基于ColDef的列样式细粒度设置

- 分别设置不同字段的宽度比例系数

```python
from plottable import ColDef

fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman',
        'ha': 'center' # 水平居中
    },
    column_definitions=[
        ColDef(name='index', width=0.5),
        ColDef(name='field1', width=2),
        ColDef(name='field2', width=0.75),
        ColDef(name='field3', width=0.75),
        ColDef(name='field4', width=0.75),
    ]
)

plt.show()
```

- 分别设置不同字段的文本对齐方式

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman'
    },
    column_definitions=[
        ColDef(name='index', textprops={'ha': 'left'}),
        ColDef(name='field4', width=1.5, textprops={'ha': 'right'}),
    ]
)

plt.show()
```

- 分别为不同字段设置数值色彩映射

```python
import matplotlib

fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman'
    },
    column_definitions=[
        ColDef(name='field3', cmap=matplotlib.cm.Reds),
        ColDef(name='field4', text_cmap=matplotlib.cm.Greens)
    ]
)

plt.show()
```

- 为字段创建分组展示

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman'
    },
    column_definitions=[
        ColDef(name='field1', group='demo group1'),
        ColDef(name='field2', group='demo group1'),
        ColDef(name='field3', group='demo group2'),
        ColDef(name='field4', group='demo group2')
    ]
)

plt.show()
```

- 为指定字段绘制列边框

```python
fig, ax = plt.subplots()
Table(
    demo_df, 
    textprops={
        'fontsize': 18,
        'fontname': 'Times New Roman',
        'ha': 'center'
    },
    column_definitions=[
        ColDef(name='index', border='left'),
        ColDef(name='field1', border='left'),
        ColDef(name='field3', border='both'),
        ColDef(name='field4', border='right')
    ]
)

plt.show()
```
